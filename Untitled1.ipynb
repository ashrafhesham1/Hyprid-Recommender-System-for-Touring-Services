```python
df_train = pd.read_csv('./data/attractions_reviews_minimal_3.csv')
```

```python
user_item_matrix = pd.pivot_table(df_train, values='rating', index='user', columns='item').fillna(0)
new_user_item_vector = pd.pivot_table(df_new_user, values='rating', index='user', columns='item')
new_user_item_vector
```

```python
ur = {308825:4,459804:5,553171:1,10038006:3}
```

```python
l =user_item_matrix.columns
ldf = pd.DataFrame(columns=l)
for i in l:
    ldf.loc['n',i] = ur[i] if i in ur.keys() else 0
ldf
```

```python
rdf = pd.DataFrame(columns=l)
for r in df_train.iterrows():
    rdf.loc[r[1]['user'],r[1]['item']] = r[1]['rating']
rdf
```

```python
rdf = rdf.fillna(0)
rdf
```
