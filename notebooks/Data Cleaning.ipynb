### hours data

```python
att_hours = att.loc[:,('location_id','hours')]
att_hours
```

```python
# extract hours data 
def extract_att_hours_data(att_hours):
    for i in range(len(att_hours)):
        hours = att_hours['hours'][i]
        if pd.isnull(hours): 
            continue
        for j in range(len(hours['week_ranges'])):
            if len(hours['week_ranges'][j]) > 0 :
                day = hours['week_ranges'][j][0]
                day_hours = str(day['open_time']) + ':' + str(day['close_time'])
                att_hours.loc[i,'day' + str(j + 1)] = day_hours 
    
    return att_hours
```

```python
att_hours = extract_att_hours_data(att_hours)
att_hours.head()
```

```python
# drop & rename
att_hours.drop(['hours'], axis = 1, inplace = True)
att_hours.rename(columns={'location_id':'id'}, inplace= True)
att_hours.head()
```

### Reviews

```python
att_reviews = att.loc[:,('location_id','reviews')]
att_reviews.head()
```

```python
att_reviews['reviews'][0][0]
```

```python
def extract_users_data(item_reviews):
    users_ids = set()
    for item in item_reviews['reviews']:
        for review in item:
            user = review['user']['username']
            users_ids.add(user)
    users_ids_df = pd.DataFrame(list(users_ids), columns = ['name'] )
    return users_ids_df

att_users = extract_users_data(att_reviews)
```

```python
att_users.head()
```

```python
def extract_reviews_data(item_reviews, users):
    reviews = {'user' : [], 'item' : [], 'title' : [], 'text': [], 'rating' : [], 'travelDate' : []}
    users_map = {v:k for k, v in users.to_dict()['name'].items()}
    for _ , item in item_reviews.iterrows():
        for review in item[1]:
            user_id = users_map[review['user']['username']]
            reviews['user'].append(user_id)
            reviews['item'].append(item[0])
            reviews['title'].append(review['title'])
            reviews['text'].append(review['text'])
            reviews['rating'].append(review['rating'])
            reviews['travelDate'].append(review['travelDate'])
            
    reviews_df = pd.DataFrame(reviews)
    return reviews_df

att_reviews_final = extract_reviews_data(att_reviews, att_users)
```

```python
att_reviews_final.head()
```

```python
#att_users.to_csv('Dataset/attractions_users.csv')
#att_reviews_final.to_csv('Dataset/attractions_users.csv.csv')
```

# Reviews minimal

```python
att_reviews_minimal = att_reviews_final.drop(['title','text'], axis = 1)

#creating timestamp column
att_reviews_minimal['timestamp'] = att_reviews_minimal['travelDate'].apply(lambda x:int(datetime.strptime(x, '%Y-%m').replace(day=1).timestamp())\
                                          if not pd.isna(x) else 0).astype(int)

#replace missing values with median
timestamp_median = att_reviews_minimal['timestamp'][att_reviews_minimal['timestamp'] != 0].median(skipna=True)
att_reviews_minimal['timestamp'] = att_reviews_minimal['timestamp'].replace(0, timestamp_median)

att_reviews_minimal.drop(['travelDate'], axis = 1, inplace= True)
att_reviews_minimal.head()
```

```python
re_hours = re.loc[:,('id','hours')]
```

```python
# extract hours data 
def extract_re_hours_data(re_hours):
    for i in range(len(re_hours)):
        hours = re_hours['hours'][i]
        if hours == []: 
            continue
        for j in range(len(hours)):
            if len(hours[j]) > 0 :
                day = hours[j][0]
                day_hours = str(day['open']) + ':' + str(day['close'])
                re_hours.loc[i,'day' + str(j + 1)] = day_hours 
    
    return re_hours
```

```python
re_hours = extract_re_hours_data(re_hours)
re_hours.head()
```
